```python
### libraries
import os
import pandas as pd
#from sklearn.feature_extraction.text import CountVectorizer
import numpy as np
#from sklearn.model_selection import train_test_split
#import tensorflow_hub as hub
import spacy
```

```python
### read in data 
sent_df = pd.read_csv('03_Outputs/04D_FULL_labeled_sampledata.csv',parse_dates=['Date'])
#sent_df = pd.read_csv('03_Outputs/04_labeled_sampledata.csv',parse_dates=['Date'])
sent_df['Year'] = sent_df['Date'].dt.year 
```

```python
sent_df.head()
```

```python
# Load pre-trained model
#30 seconds

nlp = spacy.load("en_core_web_sm")


# Define function to obtain sentence embeddings
def get_sentence_embedding(sentence):
    doc = nlp(sentence)
    return doc.vector

# Obtain sentence embeddings and create new dataframe with embeddings
embeddings = sent_df['sentences'].apply(get_sentence_embedding).apply(pd.Series)
df_embeddings = pd.concat([sent_df, embeddings], axis=1)
```
